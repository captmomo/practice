```python
from bs4 import BeautifulSoup
```

```python
html = open("post.html", encoding='utf-8').read()
soup = BeautifulSoup(html, "lxml")
```

```python
print(soup)
```

```python
soup.prettify()
```

```python
soup.find_all('span"', {"dir": "ltr"})
```

```python

main = "Full podcast will be released soon, here is a teaser clip with Lyle discussing the beloved ECA stack. Would be interested to hear people's experiences with it. I generally stay away from stimulants but I can recognize the efficacy of its use. https://www.youtube.com/watch?v=kOe1G7DSCfc&feature=youtu.be&fbclid=IwAR1DvdRlQ_rx4wm06NrSDUoB4K5vyfCWU1QnSWb0qOorEtdOc6FGvpFXwWw"
```

```python
f = open('test.md', 'w+', encoding='utf-8')
f.write(f'> {main} \n')
```

```python
for comment in soup.find_all('li'):
    parent_comment = comment.find('div', {"aria-label": "Comment"})
    if parent_comment is None:
        continue
    main_comment = parent_comment.find('span', { 'dir' : 'ltr'}).text
    main_comment = ' '.join(main_comment.split())
    main_text = f"{parent_comment.find('img').get('alt')}: {main_comment}"
    f.write(f"* {main_text} \n")
    replies = comment.find_all('div', {"aria-label": "Comment reply"})
    if len(replies) > 0:
        for reply in replies:
            reply_text = reply.find('span', { 'dir' : 'ltr'}).text.strip()
            repl_text = ' '.join(reply_text.split())
            f.write(f"  * {reply.find('img').get('alt')}: {repl_text} \n")
```
